```python
import numpy as np
import matplotlib.pyplot as plt
```

Generating random grades of 100 students below:

```python
rng= np.random.default_rng()
grades= rng.integers(0,100, size=100)
```

Calculating basic statistics on them:

```python
mean=np.mean(grades)
median= np.median(grades)
std= np.std(grades)
min= np.min(grades)
max=np.max(grades)
```

Grade Distribution:

```python
bins= [0,50,60,70,80,90,101]
grades_score=['F','E','D','C','B','A']

hist, _= np.histogram(grades, bins=bins)
```

Analysis:

```python
above_avg= np.count_nonzero(mean<grades)
below_avg= np.count_nonzero(mean>grades)
top_10_percent= np.percentile(grades, 90)
below_10_percent= np.percentile(grades, 10)
```

Final Result of full calculated value:

```python
print("--Grade Analyzer--")
print("Total Students=100\n")

print("Statistics:\n")
print(f"Mean:{mean}")
print(f"Median:{median}")
print(f"STD:{std}")
print(f"minimum score:{min}")
print(f"maximum score:{max}\n")

print("Grade Distribution:\n")
for grade,count in zip(grades_score,hist):
    print(f'{grade}:{count}\n')

print(f"Above average={above_avg}")
print(f"Below average:{below_avg}\n")

print(f"cutoff top 10%:{top_10_percent}")
print(f"cutoff bottom 10%:{below_10_percent}\n")

plt.hist(grades, bins=bins, color="green", edgecolor="black")
plt.title("Histogram of the grade distribution")
plt.xlabel("Grades")
plt.ylabel("Distribution")

plt.show()
```
